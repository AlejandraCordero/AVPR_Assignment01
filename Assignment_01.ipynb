## Objective
Design and implement a system capable of automatically detecting and counting complete circular pieces within thermographic images.

## Methodology
This implementation uses computer vision techniques including:
- Image preprocessing
- Hough Circle Transform for circle detection
- Boundary validation to ensure only complete pieces are counted
- Visual annotation for result verification

## 1. Import and Setup
Import necessary libraries for image processing and visualization.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs('output', exist_ok=True)
```

#### Image01 Detection

```python
img = cv2.imread('images/image1.png')

# Convert to grayscale for better circle detection
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply blur to reduce noise
blurred = cv2.GaussianBlur(gray, (9, 9), 0)

# Detect circles using Hough Transform
circles = cv2.HoughCircles(
    blurred, cv2.HOUGH_GRADIENT, dp=1, minDist=60, param1=50, param2=50, minRadius=100, maxRadius=145
)

# List to store only complete circles
complete_circles = []
h, w = blurred.shape

# Check if any circles were found
if circles is not None:
    for x, y, r in circles[0]:
        if (x-r > 0) and (x+r < w) and (y-r > 0) and (y+r < h):
            complete_circles.append((x, y, r))

output = img.copy()

# Draw each detected circle
for x, y, r in complete_circles:
    x, y, r = int(x), int(y), int(r)
    cv2.circle(output, (x, y), r, (0,1,0), 3)
    cv2.circle(output, (x, y), 2, (0,0,1), 3)

# Show results
print("Circles found:", len(complete_circles))
output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
cv2.imwrite('output/image01_detected.png', output)
plt.figure(figsize=(8, 8))
plt.imshow(output_rgb)
plt.show()
```
